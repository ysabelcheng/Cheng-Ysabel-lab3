```python
def relationship_status(from_member,to_member):
    follows = 0
    followed = 0
    
    
    following = (social_graph[from_member]["following"])
    for x in following:
        if (x == (to_member)):
            followed = 1
    
    
    following = (social_graph[to_member]["following"])
    for x in following:
        if (x == (from_member)):
            follows = 1
    
    
    if follows and followed:
        print ("Friends")
    elif follows or followed:
        if followed == 1:
            print ("Follows")
        if follows == 1:
            print ("Followed by")
    else:
        print("None")

        
relationship_status("@chums","@jobenilagan")
relationship_status("@jobenilagan","@joeilagan")
relationship_status("@bongolpoc","@chums")
relationship_status("@bongolpoc","@jobenilagan")
relationship_status("@bongolpoc","@chums")
relationship_status("@chums","@joeilagan")
relationship_status("@joeilagan","@chums")
relationship_status("@eeebeee", "@joeilagan")
```
